```python
import pandas as pd
import itertools as it
from finalWrapper import MyWrapper as MyLSTM
from finalWrapper2 import MyWrapper2 as MyGRU
```

## LSTM Tuning

```python
epochs = [80, 140, 200]
rnn_size = [16, 32, 64, 128]
num_layers = [1, 2, 3]
attention = [False, True]
myCombination = list(it.product(epochs, rnn_size, num_layers, attention))
LSTMSummary = pd.DataFrame(columns=['epochs', 'rnn_size',
                            'num_layers', 'attention', 'MAE', 'MSE', 'Sharpe'])
```

```python
for i in range(len(myCombination)):
    data = pd.read_csv('/Users/bowen/Desktop/H/deeplearninveststrat/Data/100_clean.csv', index_col=0)
    myepochs, myRnnSize, myNumLayers, MyAttention = myCombination[i]
   
    parameters = {'hm_epochs': myepochs,
                'n_batches': 4,
                'rnn_size': myRnnSize,
                'num_layers': myNumLayers,
                'attention': MyAttention} 
    
    lstm = MyLSTM(data, parameters)
    parameters, MAE, MSE, pnl, mysharpe = lstm.run()
    
    LSTMSummary.loc[i] = [myepochs, myRnnSize, myNumLayers, MyAttention, MAE, MSE, mysharpe]
```

```python
LSTMSummary.head()
```

```python
LSTMSummary.to_csv('LSTMSummary.csv')
```
