```python
import pandas as pd
import itertools as it
from finalWrapper import MyWrapper as MyLSTM
from finalWrapper2 import MyWrapper2 as MyGRU
from finalWrapper3 import MyWrapper3 as MyRNN
```

## LSTM Tuning

```python
epochs = [100, 150, 200, 250, 300]
rnn_size = [32]
num_layers = [1]
batchSize = [4,8,16]
learning = [0.0001, 0.001, 0.01, 0.1]
attention = [False, True]
myCombination = list(it.product(epochs, rnn_size, num_layers, batchSize, learning, attention))
LSTMSummary = pd.DataFrame(columns=['epochs', 'rnn_size',
                                    'num_layers', 'batchSize', 'learningRate', 'attention', 'MAE', 'MSE', 'Sharpe'])
```

```python
for i in range(len(myCombination)):
    data = pd.read_csv('/Users/bowen/Desktop/H/deeplearninveststrat/Data/100_clean.csv', index_col=0)
    myepochs, myRnnSize, myNumLayers, myBatchSize, myLearning, myAttention = myCombination[i]

    parameters = {'hm_epochs': myepochs,
                  'n_batches': myBatchSize,
                  'rnn_size': myRnnSize,
                  'num_layers': myNumLayers,
                  'learning_rate': myLearning,
                  'attention': myAttention}

    lstm = MyLSTM(data, parameters)
    parameters, MAE, MSE, pnl, mysharpe = lstm.run()

    LSTMSummary.loc[i] = [myepochs, myRnnSize, myNumLayers,
                          myBatchSize, myLearning, myAttention, MAE, MSE, mysharpe]
```

```python
LSTMSummary.head()
```

```python
LSTMSummary.to_csv('LSTMSummary3.csv')
```
